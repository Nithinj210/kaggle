```python
#data cleaning
# Load the newly provided datasets to examine their structure and contents
file_path_1 = '/kaggle/input/champions-league-dataset-1955-2023/UCL_AllTime_Performance_Table - UCL_Alltime_Performance_Table.csv'
file_path_2 = '/kaggle/input/champions-league-dataset-1955-2023/UCL_Finals_1955-2023 - UCL_Finals_1955-2023.csv'

# Load the datasets
ucl_data_alltime = pd.read_csv(file_path_1)
ucl_data_finals = pd.read_csv(file_path_2)

# Display the first few rows of both datasets to understand their structure
ucl_data_alltime.head(), ucl_data_finals.head()
```

```python
# Reload the dataset to start fresh
ucl_data_alltime = pd.read_csv(file_path_1)

# Check for any missing values
missing_values = ucl_data_alltime.isnull().sum()

# Display missing values
missing_values
```

```python
# Drop the '#' column
ucl_data_alltime.drop(columns=['#'], inplace=True)

# Convert relevant columns to numeric types
ucl_data_alltime['M.'] = pd.to_numeric(ucl_data_alltime['M.'], errors='coerce')
ucl_data_alltime['Dif'] = pd.to_numeric(ucl_data_alltime['Dif'], errors='coerce')
ucl_data_alltime['Pt.'] = pd.to_numeric(ucl_data_alltime['Pt.'], errors='coerce')

# Verify the data types
ucl_data_alltime.dtypes
```

```python
# Features and target variable for predicting the number of wins
features_wins = ['M.', 'Dif', 'Pt.']
target_wins = 'W'

X_wins = ucl_data_alltime[features_wins]
y_wins = ucl_data_alltime[target_wins]

# Display the first few rows of the features and target variable
X_wins.head(), y_wins.head()
```

```python
#Model Training

from sklearn.model_selection import train_test_split
# Split the data into training and testing sets
X_train_wins, X_test_wins, y_train_wins, y_test_wins = train_test_split(X_wins, y_wins, test_size=0.2, random_state=42)

# Display the shapes of the training and testing sets
X_train_wins.shape, X_test_wins.shape, y_train_wins.shape, y_test_wins.shape
```

```python
#Model Evaluation 
from sklearn.linear_model import LinearRegression
# Initialize and train the regression model
model_wins = LinearRegression()
model_wins.fit(X_train_wins, y_train_wins)

# Predict on the test set
y_pred_wins = model_wins.predict(X_test_wins)

# Display the first few predictions
y_pred_wins[:5]
```

```python
from sklearn.metrics import mean_squared_error, r2_score
# Evaluate the model
mse_wins = mean_squared_error(y_test_wins, y_pred_wins)
r2_wins = r2_score(y_test_wins, y_pred_wins)

# Print the evaluation metrics
print(f"Mean Squared Error (MSE): {mse_wins}")
print(f"R^2 Score: {r2_wins}")
```

```python
# Combine the actual and predicted values into a DataFrame for comparison
results = pd.DataFrame({'Actual Wins': y_test_wins, 'Predicted Wins': y_pred_wins})

# Display the first few rows
results.head()
```

```python
# Hypothetical performance metrics for future prediction
future_data = pd.DataFrame({
    'M.': [30],  # Number of matches
    'Dif': [15],  # Goal difference
    'Pt.': [50]   # Points accumulated
})

# Predict the number of wins
predicted_wins = model_wins.predict(future_data)

print(f"Predicted Wins: {predicted_wins[0]}")
```
